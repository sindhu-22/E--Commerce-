```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\good2\Desktop\Customerspend.csv")

numeric_cols = [
    'TotalOrders',
    'TotalQuantity',
    'AvgDiscount',
    'TotalProfit',
    'TotalRevenue'
]

for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\good2\Desktop\Customerspend.csv")

correlation = df['AvgDiscount'].corr(df['TotalProfit'])

print("Pearson Correlation:", correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = df[['AvgDiscount', 'TotalProfit']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: AvgDiscount vs TotalProfit')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet(r"C:\Users\good2\Desktop\Customerspend.parquet")

plt.figure(figsize=(8,5))
plt.hist(df["TotalRevenue"], bins=30)
plt.title("Order Value Distribution")
plt.xlabel("Total Revenue")
plt.ylabel("Count")
plt.show()
```

```python
sales = pd.read_csv(r"C:\Users\good2\Desktop\sales.csv")

sales["InvoiceDate"] = pd.to_datetime(sales["InvoiceDate"])

daily_sales = sales.groupby("InvoiceDate")["Quantity"].sum()

daily_sales.plot(figsize=(12,5))
plt.title("Daily Sales Volume")
plt.show()
```

```python
sales["Month"] = sales["InvoiceDate"].dt.month

monthly_sales = sales.groupby("Month")["Quantity"].sum()

monthly_sales.plot(kind="bar")
plt.title("Monthly Sales Volume")
plt.xlabel("Month")
plt.ylabel("Quantity Sold")
plt.show()
```
